```python
from IPython.display import clear_output\

def choose_first():
    from random import randint 
    return randint(0,1)
    

def display_board(board):
    clear_output()
    print('   |   |   ')
    print(' {} | {} | {} '.format(board[1],board[2],board[3]))
    print('   |   |   ')
    print('___ ___ ___')
    print('   |   |   ')
    print(' {} | {} | {} '.format(board[4],board[5],board[6]))
    print('   |   |   ')
    print('___ ___ ___')
    print('   |   |   ')
    print(' {} | {} | {} '.format(board[7],board[8],board[9]))
    print('   |   |   ')


def player_input(player):
    markers = [' ']*2
    while markers[player] not in ['X','O']:
        markers[player] = input('Do you want to play as X or O? ')
    
    if player== 1:
        player2 = 0
    else:
        player2 = 1
    
    if markers[player] == 'X':
        markers[player2] = 'O'
    else:
        markers[player2] = 'X'
    return markers

def place_marker(board, marker, position):
    board[position] = marker
    display_board(board)
    return board

def win_check(board, player, mark):
    win = [False, False]
    if board[1] == board[2] == board[3] == mark:
        win[player] = True
    elif board[4] == board[5] == board[6] == mark:
        win[player] = True
    elif board[7] == board[8] == board[9] == mark:
        win[player] = True
    elif board[1] == board[4] == board[7] == mark:
        win[player] = True
    elif board[2] == board[5] == board[8] == mark:
        win[player] = True
    elif board[3] == board[6] == board[9] == mark:
        win[player] = True
    elif board[1] == board[5] == board[9] == mark:
        win[player] = True
    elif board[3] == board[5] == board[7] == mark:
        win[player] = True
    else:
        pass
    return win

def space_check(board, position):
    return board[position] == ' '

def full_board_check(board):
    return ' ' not in board

def player_choice(board,player):
    next_move = int(input('Hey Player {}, what is your next move? (from 1-9)'.format(player+1)))
    
    while next_move not in range(1,10):
        next_move = int(input('Hey Player {}, what is your next move? (from 1-9)'.format(player+1)))
        
    availability = space_check(board, next_move)
    
    while availability == False:
        next_move = int(input('The position {} is not available, please enter a new position (from 1-9)'.format(next_move)))
        availability = space_check(board, next_move)
    return next_move


def replay(win):
    if win[0]==True:
        another_time = input('Player 1 won!! Do you want to play again? (yes or no)\n')
    elif win[1]==True:
        another_time = input('Player 2 won!! Do you want to play again? (yes or no)\n')
    else:
        another_time = input('There is a tie!! Do you want to play again? (yes or no)\n')
    
    if another_time.lower()=='yes':
        another_time = True
    else:
        another_time = False
    return another_time
    


#Welcome
print('Welcome to Tic-Tac Toe Game!\n')

another_time = True

while another_time == True:
    #Defining an empty board by default
    board = [' ']*10
    player = choose_first()
    player_marker = player_input(player)
    ready_to_go = input('Great! you are Player {} will play as {}. Are you ready to start (yes or no)? '.format(player+1,player_marker[player]))

    #Checking if the player is ready
    if ready_to_go.lower() == 'no':
        print('Oh, sorry! we can try it at another time')
        break
    else:
        player = 1
        win = [False,False]
        while (full_board_check(board) == False) and (win == [False, False]):
            if player == 1:
                player = 0
            else:
                player = 1
                
            position = player_choice(board,player)
            board = place_marker(board,player_marker[player],position)
            win = win_check(board,player,player_marker[player])
    
    
    #After the game ends
    another_time = replay(win)
print('Bye Bye')
```
